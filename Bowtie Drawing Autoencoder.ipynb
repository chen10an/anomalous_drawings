# Bowtie Drawing Autoencoder

```python
# libraries
import numpy as np
from tqdm import tqdm

# files
import utils
from model import Autoencoder

# reproducibility
np.random.seed(0)
```

## Google Quick, Draw! Dataset
Source: https://github.com/googlecreativelab/quickdraw-dataset

```python
data = np.load('full%2Fnumpy_bitmap%2Fbowtie.npy')

print("data: {} with shape {}, dtype {}, and values in range {}".format(type(data), data.shape, data.dtype,
                                                          [np.amin(data), np.amax(data)]))
```

```python
# rescale data to range [0,1]
data = data/255

print("values in data are now in range {}".format([np.amin(data), np.amax(data)]))
```

```python
utils.plotImgs(np.reshape(data, (data.shape[0],28,28)), 100, 10, cmap='gray_r')
```

```python
np.random.shuffle(data)
```

```python
# split data into training (80%) and testing (20%) after shuffling
eightyPercent = int(np.rint(data.shape[0]*0.8))
X_train = data[:eightyPercent]
X_test = data[eightyPercent:]

X_train.shape, X_test.shape
```

## Autoencoder

```python
model = Autoencoder(X_train, X_test, epochs=50, early_stop=5, log_file='bowtie_logs')
model.train(model_file='bowtie_model/model.ckpt')
```

## Plot results

```python
print("\nTesting model on its input (train data)...")
pred_train = model.sess.run(model.decoder, feed_dict={model.X: X_train})
# compare
print("Showing sample images from the model's reconstruction...")
utils.plotImgs(np.reshape(pred_train, (pred_train.shape[0],28,28)), 25, 5, figsize=(5,5), cmap='gray_r')
print("Showing sample images from the original data...")
utils.plotImgs(np.reshape(X_train, (X_train.shape[0],28,28)), 25, 5, figsize=(5,5), cmap='gray_r')

print("\nTesting model on test data...")
pred_test = model.sess.run(model.decoder, feed_dict={model.X: X_test})
# compare
print("Showing sample images from the model's reconstruction...")
utils.plotImgs(np.reshape(pred_test, (pred_test.shape[0],28,28)), 25, 5, figsize=(5,5), cmap='gray_r')
print("Showing sample images from the original data...")
utils.plotImgs(np.reshape(X_test, (X_test.shape[0],28,28)), 25, 5, figsize=(5,5), cmap='gray_r')
```

## Anomaly Detection
Inspiration:
* http://colinmorris.github.io/blog/bad_flamingos
* https://medium.com/@curiousily/credit-card-fraud-detection-using-autoencoders-in-keras-tensorflow-for-hackers-part-vii-20e0c85301bd

As we can see on the plots above, the autoencoder has learned the basic shape of a conventional/normal bowtie, or the shape that is most common in the dataset. Even when the true image is not a normal bowtie, the model tries to reconstruct a normal bowtie. For example:

```python
# example 10 (abnormal) from the training data
print("reconstruction:")
utils.plotImgs(np.reshape(pred_train[10], (1,28,28)), 1, 1, figsize=(3,3), cmap='gray_r')

print("true image:")
utils.plotImgs(np.reshape(X_train[10], (1,28,28)), 1, 1, figsize=(3,3), cmap='gray_r')
```

As a result, we can assume that the reconstruction error for examples like the one above would be relatively large: the true image is abnormal, but the model still tries to reconstruct a normal bowtie, thus deviating from the true image.

```python
# example 10 (abnormal) from the training data
c10 = model.sess.run(model.cost, {model.X:np.expand_dims(X_train[10], axis=0)})

# examples 2, 3, 4 (normal) from the training data
c2 = model.sess.run(model.cost, {model.X:np.expand_dims(X_train[2], axis=0)})
c3 = model.sess.run(model.cost, {model.X:np.expand_dims(X_train[3], axis=0)})
c4 = model.sess.run(model.cost, {model.X:np.expand_dims(X_train[4], axis=0)})

print("error for example 10 (true image is abnormal): {}".format(c10))
print("error for example 2 (true image is normal): {}".format(c2))
print("error for example 3 (true image is normal): {}".format(c3))
print("error for example 4 (true image is normal): {}".format(c4))
```

When the true image is abnormal, the reconstruction error is relatively large. We can flip this conclusion: when the reconstruction error is relatively large, the true image is abnormal. This can be used find the abnormal examples in the true images. In other words, we can use the autoencoder's reconstruction error to find the bowtie drawings in the dataset that deviate from how people normally draw bowties.

```python
# try different thresholds: if error > threshold then the true image counts as abnormal
threshold = 0.07

# get abnormal images from the test data
abnormal = None
abnormal_dex = []
for i in tqdm(range(X_test.shape[0])):
    expanded = np.expand_dims(X_test[i], axis=0)
    c = model.sess.run(model.cost, {model.X:expanded})
    if c > threshold:
        if abnormal is None:
            abnormal = expanded
        else:
            abnormal = np.concatenate((abnormal,expanded))

print("number of abnormal images: {}\ntotal number of images: {}".format(abnormal.shape[0], X_test.shape[0]))
print("abnormal/(all images) = {}".format(abnormal.shape[0]/X_test.shape[0]))
```

```python
# plot some of these abnormal bowties
# if there are too many normal-looking bowties, then increase the threshold
utils.plotImgs(np.reshape(abnormal[:100], (100,28,28)), 100, 10, cmap='gray_r')
utils.plotImgs(np.reshape(abnormal[100:200], (100,28,28)), 100, 10, cmap='gray_r')
utils.plotImgs(np.reshape(abnormal[200:300], (100,28,28)), 100, 10, cmap='gray_r')
utils.plotImgs(np.reshape(abnormal[300:400], (100,28,28)), 100, 10, cmap='gray_r')
utils.plotImgs(np.reshape(abnormal[400:500], (100,28,28)), 100, 10, cmap='gray_r')
```

Just like this, we've found some abnormal bowtie drawings: some are not bowtie drawings at all, others are are less conventional bowtie drawings (e.g. bowtie with the string attached)
